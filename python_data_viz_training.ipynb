```python
seattle_weather = pd.read_csv("seattleWeather_2016-2017.csv")
```

```python
print(seattle_weather.head(5))
```

```python
# Adding data to the plot
fig, ax = plt.subplots()
ax.plot(seattle_weather["DATE"],seattle_weather["TMIN"])
plt.show()
```

```python
# Adding another series to the plot
fig, ax = plt.subplots()
ax.plot(seattle_weather["DATE"],seattle_weather["TMIN"])
ax.plot(seattle_weather["DATE"],seattle_weather["TMAX"])
plt.show()
```

```python
seattle_weather = seattle_weather[:5]

# Adding markers,color and linestyles to the plot
fig, ax = plt.subplots()
ax.plot(seattle_weather["DATE"],seattle_weather["TMIN"],marker="o",linestyle='-.',color='#FF0000') # hex code for Red
ax.plot(seattle_weather["DATE"],seattle_weather["TMAX"],marker="v",linestyle='--',color='#A52A2A')  #RGB code for Brown
plt.show()

# Even works with color = 'blue' or 'red'  etc
```

```python
# For plots without Lines joining the points, marker as X and Color as 'Magenta'
fig, ax = plt.subplots()
ax.plot(seattle_weather["DATE"],seattle_weather["PRCP"],marker="x",linestyle='None',color='magenta')
plt.show()
```

```python
# For labeling the axes of the plots and Title of graph, Labels
fig, ax = plt.subplots()
ax.plot(seattle_weather["DATE"],seattle_weather["TMIN"],color='#FF0000',label='Min Temp') # hex code for Red
ax.plot(seattle_weather["DATE"],seattle_weather["TMAX"],color='#A52A2A',label='Max Temp')  #RGB code for Brown

ax.set_xlabel("Time-days")

ax.set_ylabel("Min and Max Temperature")

ax.set_title("Weather patterns in Seattle")

plt.legend()
plt.show()
```
